```python
df = df.drop(columns = 'Not_Useful_Column')
```

```python
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r'[/|-]', '', regex=True)
df
```

```python
#Reminde: run each line separately

#df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','') #replaces things like ./'\ which are not within the collumn

#df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

#df["Phone_Number"].apply(lambda x: str(x))

df['Phone_Number'] = df["Phone_Number"].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:10]) #for this to work everything in code should be in string , previously we had Na

#df['Phone_Number'] = df['Phone_Number'].fillna('').astype(str)

#df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x if len(x) != 10 else f"{x[0:3]}-{x[3:6]}-{x[6:10]}"


#df['Phone_Number'] = df['Phone_Number'].Replace('Na','',regex = True)

#df["Phone_Number"] = df["Phone_Number"].str.replace(r'[Na]', '', regex=True)  #vip code

df







```

```python
df['Phone_Number'] = df['Phone_Number'].replace('--', '')  #here we didnot call string replace , so it handled in without string
```

```python
df
```

```python
df[['street_address', 'state', 'zipcode']] = df['Address'].str.split(',', n=2, expand=True)
df
```

```python
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('Yes','Y')

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
```

```python
df
```

```python
#df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('N/a','')
#df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('NaN','') #this dont work coz there is nothing to replace

df['Do_Not_Contact'] = df['Do_Not_Contact'].fillna('')

```

```python
df
```

```python
for x in df.index:  #Loops through all the row indices in the DataFrame.
    if df.loc[x,"Do_Not_Contact"] == 'Y': #For each row (identified by x), checks if the value in the column "Do_Not_Contact" is equal to 'Y'.
        df.drop(x,inplace = True) #Removes the row with index x from the DataFrame if the condition is met

df
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"] == '':
        df.drop(x,inplace = True)

df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```
